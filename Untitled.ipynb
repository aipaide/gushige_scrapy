```python
# 针对诗歌的问题模板
# author="李白"
# title="赠王伦"
# theme="离别"
# words=12
# time="唐朝"
# 一般
question_shige_general = [
    '请做一首诗',
    '请做一首古诗'
]

# 作者
question_shige_author = [
    '请写一首{author}的诗',
]

# title
question_shige_title = [
    '请以“{title}”为题，写一首诗。',
]

# time
question_shige_title = [
    '请写一首{time}时代的一首诗。',
]

# 主题
question_shige_themes = [
    '请写一首关于“{theme}”的诗'
]

# 字数
question_shige_words = [
     '请写一首不少于{words}字的诗'
]


question_templates = question_shige_general + question_shige_author + question_shige_title + question_shige_themes+question_shige_words
```

```python
title='sdf'
question_templates
```

```python
import jsonlines
story_qa_data=[]
f = open ('gushici.jsonl', "r")
with jsonlines.open('gushici.jsonl') as reader:
    for obj in reader:
        title = obj['title']
        author = obj['author']
        time = obj['time']
        words = len(obj['content'])
        themes = obj['themes']
        content = obj['content']
        templates = {
            'title':title,
            'author':author,
            'time': time,
            'words': words,
            'theme': theme
        }
        for q in question_shige_general:
                res = {
                      "instruction": q.format(**templates),
                      "input": "",
                      "output": content
                      }
                story_qa_data.append(res)
                
        if author is not None:
            for q in question_shige_author:
                res = {
                      "instruction": q.format(**templates),
                      "input": "",
                      "output": content
                      }
                story_qa_data.append(res)
        if title is not None:
            for q in question_shige_title:
                res = {
                      "instruction": q.format(**templates),
                      "input": "",
                      "output": content
                      }
                story_qa_data.append(res)
                
        for theme in themes:
            templates['theme'] = theme
            for q in question_shige_themes:
                res = {
                      "instruction": q.format(**templates),
                      "input": "",
                      "output": content
                      }
                story_qa_data.append(res)
        if words > 0 :
            for q in question_shige_words:
                res = {
                      "instruction": q.format(**templates),
                      "input": "",
                      "output": content
                      }
                story_qa_data.append(res)
```
